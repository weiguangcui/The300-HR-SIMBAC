```python
def get_cent(SNAME,overden):
    path='/home2/weiguang/The300/data/catalogues/AHF/'
    outd=np.zeros((324,7),dtype=np.float64)
    cc=np.array([500000,500000,500000])
    for i in np.arange(1,325):
        outd[i-1,0]=i
        if overden is None:
            ahfd=np.loadtxt(path+'%s/NewMDCLUSTER_%04d/%s-NewMDCLUSTER_%04d.snap_128.z0.000.AHF_halos'%(SNAME,i,SNAME,i),usecols=(0,1,3,5,6,7,11))
        else:
            ahfd=np.loadtxt(path+'%s_%s/NewMDCLUSTER_%03d/%s-NewMDCLUSTER_%03d.snap_128.z0.000.AHF_halos'%(SNAME,overden,i,SNAME,i),usecols=(0,1,3,5,6,7,11))
            
        if overden=='R500c':
            ml=1e14
        else:
            ml=3e14
        rr=np.sqrt(np.sum((ahfd[:,3:6]-cc)**2,axis=1))
        ids=np.where((rr<2000)&(ahfd[:,1]<1)&(ahfd[:,2]>ml))[0]  #need rr<3mpc/h for Rvir
        if len(ids)==1:
            outd[i-1,1]=ahfd[ids,0][0]
            outd[i-1,2:]=ahfd[ids,2:][0]
        elif len(ids)==0:
            ids=np.where((rr<3000)&(ahfd[:,1]<1)&(ahfd[:,2]>ml))[0]  #need rr<3mpc/h for Rvir cl--47
            if len(ids)==1:
                outd[i-1,1]=ahfd[ids,0][0]
                outd[i-1,2:]=ahfd[ids,2:][0]
            else:
                print("!! increasing rr not work!!!!",i,ids)
        elif len(ids)>1:
            ids=np.where((rr<1000)&(ahfd[:,1]<1)&(ahfd[:,2]>ml))[0]  #reduce rr for multiple match
            if len(ids)==1:
                outd[i-1,1]=ahfd[ids,0][0]
                outd[i-1,2:]=ahfd[ids,2:][0]
            else:
                print("!! decreasing rr not work!!!!",i,ids) 
                print('!! I use the one with more mass')
                idn=ids[np.argmax(ahfd[ids,2])]
                outd[i-1,1]=ahfd[idn,0][0]
                outd[i-1,2:]=ahfd[idn,2:][0]
        else:
            print("!! no halo found!!",i,ids)
    if overden is None: #R200c
        np.savetxt('%s_R200c_snaps_128-center-cluster.txt'%SNAME,outd, fmt='%d %d'+ ' %.5e'*5,header='[0]region ID; [1]AHF ID; [2]M_halo Msun/h; [3-5]Xc,Yc,Zc kpc/h; [6]R_halo kpc/h')
    else:
        np.savetxt('%s_%s_snaps_128-center-cluster.txt'%(SNAME,overden),outd, fmt='%d %d'+ ' %.5e'*5,header='[0]region ID; [1]AHF ID; [2]M_halo Msun/h; [3-5]Xc,Yc,Zc kpc/h; [6]R_halo kpc/h')
```
